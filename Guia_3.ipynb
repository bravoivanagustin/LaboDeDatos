```python
import seaborn as sns
import seaborn.objects as so
import pandas as pd
import matplotlib.pyplot as plt
from gapminder import gapminder
```

```python
display(gapminder)
```

```python
pxc = gapminder.groupby("continent")["country"].unique()
pxc
```

```python
(
    so.Plot(x = pxc.index, y = pxc)
    .add(so.Bar())
)
```

```python
(
    so.Plot(data = gapminder, x = "continent")
    .add(so.Bar(), so.Hist())
)
```

```python
sns.boxplot(gapminder, x = "continent", y = "lifeExp", order = sorted(gapminder["continent"].unique()))
```

```python
arg = gapminder[gapminder["country"] == "Argentina"]
```

```python
(
    so.Plot(data = arg, x = "year", y = "lifeExp")
    .add(so.Line())
)
```

```python
ame = gapminder[gapminder["continent"] == "Americas"]
asia = gapminder[gapminder["continent"] == "Asia"]
```

```python
(
    so.Plot(data = ame, x = "year", y = "lifeExp", color = "country")
    .add(so.Line())
)
```

```python
(
    so.Plot(data = ame, x = "year", y = "lifeExp")
    .add(so.Lines(color = "gray"), group = "country")
    .add(so.Line(), so.PolyFit(1))
)
```

```python
(
    so.Plot(data = asia, x = "year", y = "lifeExp", color = "country")
    .add(so.Line())
)
```

```python
(
    so.Plot(gapminder , x = "year", y = "lifeExp")
    .add(so.Lines(color = "lightgray"), group = "country")
    .add(so.Line(), so.PolyFit(1))
    .facet("continent", wrap = 3)
)
```

```python
penguins = sns.load_dataset("penguins")
```

```python
(
so.Plot(data = penguins , x = "bill_length_mm" , y = "bill_depth_mm" ,
edgewidth = "body_mass_g" , marker = "species" ,
linestyle = "island" , color = "species")
.add(so.Dot(color = ".8" ) , edgecolor = "sex")
)
```

```python
(
    so.Plot(data = penguins, x = "bill_length_mm", y = "bill_depth_mm", color = "species")
    .add(so.Dot())
)
```

```python
(
    so.Plot(data = penguins, x = "bill_length_mm", y = "bill_depth_mm", color = "sex")
    .add(so.Dot())
)
```

```python
(
    so.Plot(data = penguins, x = "bill_length_mm", y = "bill_depth_mm", color = "sex")
    .add(so.Dot())
    .facet("species")
)
```
